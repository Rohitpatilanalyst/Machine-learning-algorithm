```python
import pandas as pd  
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
 
```

```python
link = "G:/Machine learning with python/nba_logreg.csv"
data = pd.read_csv(link)
#specification of input and output variables 
X = data.iloc[:, 16:20].values  
y = data.iloc[:, 20].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
data.head()
```

```python
 #Evaluate the accuracy using 

svc_lin = SVC(kernel='linear')
svc_lin.fit(X_train,y_train)
svc_lin_pred=svc_lin.predict(X_test)
lin_acc=accuracy_score(y_test,svc_lin_pred)

svc_sig = SVC(kernel='sigmoid')
svc_sig.fit(X_train,y_train)
svc_sig_pred=svc_sig.predict(X_test)
sig_acc=accuracy_score(y_test,svc_sig_pred)

svc_poly = SVC(kernel='poly')
svc_poly.fit(X_train,y_train)
svc_poly_pred=svc_poly.predict(X_test)
poly_acc=accuracy_score(y_test,svc_poly_pred)

svc_rad = SVC()
svc_rad.fit(X_train,y_train)
svc_rad_pred=svc_rad.predict(X_test)
rad_acc=accuracy_score(y_test,svc_rad_pred)

accuracy=[lin_acc,sig_acc,poly_acc,rad_acc]

accuracy
 
```

```python
#Evaluate the accuracy using 
#m = np.array([])
m=np.zeros((1,4))
ratio=[0.4,0.5,0.6,0.7,0.8,0.9]
for i in ratio:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=i, random_state=0) 

    svc_lin = SVC(kernel='linear')
    svc_lin.fit(X_train,y_train)
    svc_lin_pred=svc_lin.predict(X_test)
    lin_acc=accuracy_score(y_test,svc_lin_pred)

    svc_sig = SVC(kernel='sigmoid')
    svc_sig.fit(X_train,y_train)
    svc_sig_pred=svc_sig.predict(X_test)
    sig_acc=accuracy_score(y_test,svc_sig_pred)

    svc_poly = SVC(kernel='poly')
    svc_poly.fit(X_train,y_train)
    svc_poly_pred=svc_poly.predict(X_test)
    poly_acc=accuracy_score(y_test,svc_poly_pred)

    svc_rad = SVC()
    svc_rad.fit(X_train,y_train)
    svc_rad_pred=svc_rad.predict(X_test)
    rad_acc=accuracy_score(y_test,svc_rad_pred)

    accuracy=[lin_acc,sig_acc,poly_acc,rad_acc]

    m=np.append(m, [accuracy], axis=0)
#np.shape(m) 
 
```

```python
x=[0.4,0.5,0.6,0.7,0.8,0.9]
y1=m[1:7,0]
y2=m[1:7,1]
y3=m[1:7,2]
y4=m[1:7,3]
 
```

```python
# multiple line plot
import matplotlib.pyplot as plt

plt.plot( x, y1,  marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4,label="linear")
plt.plot( x, y2,  marker='', color='olive', linewidth=2,label="sigmoid")
plt.plot( x, y3,  marker='', color='red', linewidth=2, linestyle='dashed', label="poly")
plt.plot( x, y4,  marker='', color='green', linewidth=2, linestyle='dashed', label="radial")
plt.legend()
```
